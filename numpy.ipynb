```python
files = ['assets/data1.tsv', 'assets/data2.tsv', 'assets/data3.tsv', 'assets/data4.tsv']

for f in files:
    data = np.loadtxt(f, delimiter='\t')
    x = data[:, 0] #første kolonne (x-værdier)
    y = data[:, 1] #anden kolonne (y-værdier)

    x_mean = np.mean(x)
    y_mean = np.mean(y)

  
    print(f"{f}: x-gennemsnit = {x_mean:.2f}, y-gennemsnit = {y_mean:.2f}")
```
