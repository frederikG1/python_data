```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('../assets/police.csv')
```

```python
total = len(df)
print(f"Total antal forbrydelser: {total}")
```

```python
categories = df["Incident Category"].unique()
print(f"\nAntal kategorier: {len(categories)}")
print("\nKategorier:")
print(categories)
```

```python
crimes_per_category = df["Incident Category"].value_counts()
print("\nAntal forbrydelser per kategori:")
print(crimes_per_category)
```

```python
plt.bar(crimes_per_category.index, crimes_per_category.values)
plt.xticks(rotation=90)  # rotation så kategorierne kan læses
plt.tight_layout()
plt.show()
```

```python
crimes_most_per_year = df["Incident Year"].value_counts().sort_index()
print(f"\nÅr med flest forbrydelser: {crimes_most_per_year.idxmax()}")
```

```python
crimes_least_pear_year = df["Incident Year"].value_counts().sort_index()
print(f"\nÅr med færreste forbrydelser: {crimes_least_pear_year.idxmin()}")
```

```python
plt.bar(crimes_most_per_year.index, crimes_most_per_year)
plt.show()
```
